## Pour le deuxième tour

```python
df = pd.read_excel('Presidentielle_2017_Resultats_Tour_1_c.xls', sheetname=2, skiprows=3)
```

```python
usecols = ['Code du département', 'Libellé du département', '% Abs/Ins', 'Nom', 'Prénom', 'Voix',
       '% Voix/Ins', '% Voix/Exp', 'Sexe.1', 'Nom.1', 'Prénom.1', 'Voix.1',
       '% Voix/Ins.1', '% Voix/Exp.1', 'Nom.2', 'Prénom.2', 'Voix.2',
       '% Voix/Ins.2', '% Voix/Exp.2', 'Nom.3', 'Prénom.3', 'Voix.3',
       '% Voix/Ins.3', '% Voix/Exp.3', 'Nom.4', 'Prénom.4', 'Voix.4',
       '% Voix/Ins.4', '% Voix/Exp.4', 'Nom.5', 'Prénom.5', 'Voix.5',
       '% Voix/Ins.5', '% Voix/Exp.5']
```

```python
df = df[usecols]
```

```python
df.columns = ['id', 'title', '% Abs/Ins', 'Nom',
       'Prénom', 'Voix', '% Voix/Ins', '% Voix/Exp', 'Sexe.1', 'Nom.1',
       'Prénom.1', 'Voix.1', '% Voix/Ins.1', '% Voix/Exp.1', 'Nom.2',
       'Prénom.2', 'Voix.2', '% Voix/Ins.2', '% Voix/Exp.2', 'Nom.3',
       'Prénom.3', 'Voix.3', '% Voix/Ins.3', '% Voix/Exp.3', 'Nom.4',
       'Prénom.4', 'Voix.4', '% Voix/Ins.4', '% Voix/Exp.4', 'Nom.5',
       'Prénom.5', 'Voix.5', '% Voix/Ins.5', '% Voix/Exp.5']
```

```python
df.head()
```

```python
keepCols = [col for col in df.columns if not col.find('Voix/Ins') > 0 and col[:4] != 'Voix']
len(keepCols)
```

```python
df = df[keepCols]
```

```python
df.head()
```

```python
newCols = []
for col in df.columns:
    newCols.append(col.replace('.', '_').replace('é','e').replace('% Voix/Exp', 'Score').replace('% Abs/Ins', 'Abstention'))
newCols
```

```python
df.columns = newCols
```

```python
del df['Sexe_1']
```

```python
def toTitle(value):
    return value.title()
nom_cols = [col for col in df.columns if col.find('Nom') >= 0]
for c in nom_cols:
    df[c] = df[c].apply(toTitle)
```
